## Challenge 1
There is a box with 10 white balls, 12 red balls and 8 black balls. Calculate the probability of:
* Taking a white ball out.
* Taking a white ball on first extraction and a black ball out on second extraction without replacing the first one.
* Taking a white ball on first extraction and a black ball out on second extraction replacing the first one.
* Taking 3 balls and them being all different colours.

```python
pw = 10/(10+12+8)
print(f"The probability of taking a white ball out: {pw:.2%}")
pwb = pw * 8/(8+12+9)
print(f"The probability of taking a white ball on first extraction and a black ball out on second extraction without replacing the first one: {pwb:.2%}")
pwbr = pw * 8/(8+12+10)
print(f"The probability of taking a white ball on first extraction and a black ball out on second extraction replacing the first one: {pwbr:.2%}")
p3b = 6*((10/30)*(12/29)*(8/28))
print(f"The probability of taking 3 balls and them being all different colours: {p3b:.2%}")
```

A and B are events of a probability such that $P(A) = 0.3$, $P(B) = 0.6$ and $P(A \cap B) = 0.1$

Which of the following statements are true?  
**HINT: draw a big Venn Diagram by your side**  
* $P(\overline{A}) = 0.7$
* $P(A \cup B) = 0.6$
* $P(A \cap \overline{B}) = 0.2$
* $P(A \cup \overline{A}) = 1$  
* $P(A \cap (B \cup \overline{B})) = 0.4$
* $P(\overline{A} \cap \overline{B}) = 0.3$
* $P(\overline{A \cap B}) = 0.9$

```python
# --- CHALLENGE 3: SET THEORY AND PROBABILITY ---
# Given:
# P(A) = 0.3
# P(B) = 0.6
# P(AnB) = 0.1 (Intersection)

# 1. P(not A) = 1 - P(A) = 1 - 0.3 = 0.7
# STATEMENT: P(not A) = 0.7 
# VERDICT: TRUE

# 2. P(A u B) = P(A) + P(B) - P(AnB) = 0.3 + 0.6 - 0.1 = 0.8
# STATEMENT: P(A u B) = 0.6 
# VERDICT: FALSE (Actual is 0.8)

# 3. P(A and not B) = P(A) - P(AnB) = 0.3 - 0.1 = 0.2
# STATEMENT: P(A and not B) = 0.2 
# VERDICT: TRUE

# 4. P(A u not A) = P(Sample Space) = 1
# STATEMENT: P(A u not A) = 1 
# VERDICT: TRUE

# 5. P(A n (B u not B)) = P(A n 1) = P(A) = 0.3
# STATEMENT: P(A n (B u not B)) = 0.4 
# VERDICT: FALSE (Actual is 0.3)

# 6. P(not A n not B) = 1 - P(A u B) = 1 - 0.8 = 0.2
# STATEMENT: P(not A n not B) = 0.3 
# VERDICT: FALSE (Actual is 0.2)

# 7. P(not (AnB)) = 1 - P(AnB) = 1 - 0.1 = 0.9
# STATEMENT: P(not (AnB)) = 0.9 
# VERDICT: TRUE

# --- SUMMARY OF TRUE STATEMENTS ---
# * P(not A) = 0.7
# * P(A and not B) = 0.2
# * P(A u not A) = 1
# * P(not (AnB)) = 0.9
```

## Challenge 4

 * If someone flies to China, the probability of getting Covid is 1%  
 * If someone flies to Italy, the probability of getting Covid is 20%  

I roll 1 dice with numbers 1, 2, 3, 4, 5, 6:  
 * If 1: I fly to Italy
 * If not 1: I fly to China

During my trip, I got Covid. What is the probability that I flew to China?

```python
pchina = 5/6 * 0.01
pitaly = 1/6 * 0.2
pcovid = pchina + pitaly
pflychina = pchina / pcovid
print(f"The probability that I flew to China is: {pflychina:.2%}")
```

## BONUS

You roll 2 dice. What is the probability that the sum of the numbers is 7? And 8? And 9? Then simulate 10000 times and draw a countplot.

your explanation here

```python
import random
import seaborn as sns
import matplotlib.pyplot as plt

n_simulations = 10000
sums = []

for _ in range(n_simulations):
    die1 = random.randint(1, 6)
    die2 = random.randint(1, 6)
    sums.append(die1 + die2)

sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))

ax = sns.countplot(x=sums, hue=sums, palette="viridis", legend=False)

plt.title(f"Distribution of Sums from {n_simulations} Dice Rolls", fontsize=15)
plt.xlabel("Sum of Two Dice", fontsize=12)
plt.ylabel("Frequency", fontsize=12)

total = len(sums)
for p in ax.patches:
    percentage = f'{100 * p.get_height() / total:.1f}%'
    ax.annotate(percentage, (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='baseline', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.show()
```
